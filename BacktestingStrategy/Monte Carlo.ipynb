```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
capital = 100
bet_size = 3

prob_win = 0.5
prob_loss = (1 - prob_win)

win_reward = 1
loss_reward = -1

```

```python
fig, ax = plt.subplots(figsize=(16, 8))

last_result = []

for j in range(1000): 
    equity_curve = [capital]
    
    for i in range(1000):
        if prob_loss < random.random():
            equity_curve += [equity_curve[-1] + win_reward * bet_size]
        else:
            equity_curve += [equity_curve[-1] + loss_reward * bet_size]
    last_result += [equity_curve[-1] - capital]
    ax.plot(equity_curve)
```

```python
import numpy as np
np.mean(last_result)
```

# Simulando um efeito Martingale 

```python
capital = 1000
bet_size = 10
bet_size_start = 10

prob_win = 0.5
prob_loss = (1 - prob_win)

win_reward = 1
loss_reward = -1

```

```python
fig, ax = plt.subplots(figsize=(16, 8))

last_result = []

for j in range(1000): 
    bet_size = 10
    equity_curve = [capital]
    for i in range(100):
        if prob_loss < random.random():
            equity_curve += [equity_curve[-1] + win_reward * bet_size]
            bet_size = bet_size_start
        else:
            equity_curve += [equity_curve[-1] + loss_reward * bet_size]
            bet_size = bet_size * 2
        if equity_curve[-1] <= 0:
            break
    last_result += [equity_curve[-1] - capital]
    ax.plot(equity_curve)
```
