### 2. ParquetFiles

```python
# library pra interagir com formato Parquet
import pyarrow as pa
import pyarrow.parquet as pq
```

### Writer

```python
t0 = time.time()
table = pa.Table.from_pandas(df_intraday)
writer = pq.ParquetWriter("petr4.parquet", table.schema)     

writer.write_table(table)
writer.close()
t1 = time.time()
print(t1 - t0)
```

### Reader

```python
pq.ParquetFile("petr4.parquet").read().to_pandas()
```
